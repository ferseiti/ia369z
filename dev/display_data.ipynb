```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
even_dir = 'Espinho_sub_even/recon'
odd_dir = 'Espinho_sub_odd/recon'

odd = os.listdir(odd_dir)
even = os.listdir(even_dir)
```

```python
length = 1
print(format(length, '03'))
```

```python
for e in sorted(even, reverse=True):
    old_number = format(int(e.replace('slice_','').replace('.b', '')), '03')
    new_number = format(int(e.replace('slice_','').replace('.b', '')) *2, '03')
    print(e, old_number, new_number, 'slice_{}.b'.format(new_number))
    os.rename('{}/{}'.format(even_dir, e), '{}/slice_{}.b'.format(even_dir, new_number))
```

```python
for o in sorted(odd, reverse=True):
    old_number = format(int(o.replace('slice_','').replace('.b', '')), '03')
    new_number = format(int(o.replace('slice_','').replace('.b', '')) *2 + 1, '03')
    print(o, old_number, new_number, 'slice_{}.b'.format(new_number))
    os.rename('{}/{}'.format(odd_dir, o), '{}/slice_{}.b'.format(odd_dir, new_number))
```
